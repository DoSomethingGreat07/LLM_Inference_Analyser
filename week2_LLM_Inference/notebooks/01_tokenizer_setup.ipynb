# 01 Tokenizer Setup

This notebook is Step 1 of the Week 2 LLM Inference Analyzer project.

Goal for this notebook:
- load the tokenizer only
- tokenize one simple sentence
- inspect characters, words, token IDs, token pieces, token count, and decoded text

Important:
- we are **not** loading the full model
- we are **not** generating text
- we are focusing only on tokenization

## Cell 1: Add the `src` folder to Python's import path

This lets the notebook import our reusable helper file from `src/tokenizer_utils.py`.

```python
from pathlib import Path
import sys

project_root = Path.cwd().parent
src_path = project_root / "src"

if str(src_path) not in sys.path:
    sys.path.append(str(src_path))

print(f"Project root: {project_root}")
print(f"Source path added: {src_path}")
```

## Cell 2: Import the helper functions

We will reuse functions from `tokenizer_utils.py` instead of writing everything directly in the notebook.

```python
from tokenizer_utils import (
    load_tokenizer,
    tokenize_text,
    count_characters,
    count_words,
)
```

## Cell 3: Load the open-source tokenizer

This downloads or loads the tokenizer for `Qwen/Qwen2.5-1.5B-Instruct`.

Only the tokenizer is loaded here, not the full LLM weights.

```python
tokenizer = load_tokenizer()
print(tokenizer)
```

## Cell 4: Define the sample sentence

This is the exact sentence we want to analyze in Step 1.

```python
text = "Hello, I am learning LLM infrastructure."
print(text)
```

## Cell 5: Check character count and word count

- `count_characters(text)` counts every character, including spaces and punctuation.
- `count_words(text)` counts words by splitting on spaces.

```python
print("Original text:", text)
print("Character count:", count_characters(text))
print("Word count:", count_words(text))
```

## Cell 6: Tokenize the sentence

The helper function returns a dictionary with:
- token IDs
- token pieces
- token count
- decoded text

```python
result = tokenize_text(text)
result
```

## Cell 7: Print the tokenization results clearly

This makes the output easier to read for a first experiment.

```python
print("Original text:", result["original_text"])
print("Character count:", result["character_count"])
print("Word count:", result["word_count"])
print("Token IDs:", result["token_ids"])
print("Token pieces:", result["token_pieces"])
print("Token count:", result["token_count"])
print("Decoded text:", result["decoded_text"])
```

## Cell 8: What this shows

This first experiment helps us see that:
- text is not processed as whole words by default
- a tokenizer breaks text into smaller units called tokens
- token count is different from character count and word count
- decoding helps us verify how tokens map back to text
